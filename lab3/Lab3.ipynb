```python
#laboratoria
from abc import ABC, abstractmethod
class Samochód:
  def __init__(self, marka, model, rok_produkcji):
    self.marka = marka
    self.model = model
    self.rok_produkcji = rok_produkcji

  def wypisz(self):
    print(f"Marka: {self.marka}")
    print(f"Model: {self.model}")
    print(f"Rok produkcji: {self.rok_produkcji}")


wozik = Samochód("Toyota", "Corolla", 2020)
wozik.wypisz()

class Pojazd(Samochód):
  def __init__(self, marka, model, rok_produkcji, kolor):
    super().__init__(marka, model, rok_produkcji)
    self.kolor = kolor

  class Zwierzak(ABC):
    @abstractmethod
    def dźwięk(self):
      pass

class Kot(Zwierzak):
  def dźwięk(self):
    print("Miau")

class Pies(Zwierzak):
  def dźwięk(self):
    print("Hau")
```

```python
#z2
import sys
import re
class Employee:
  def __init__(self, name,surname,age, salary):
    self.name = name
    self.surname = surname
    self.age = age
    self.salary = salary
  def display(self):
    print(f"Name: {self.name}")
    print(f"Surname: {self.surname}")
    print(f"Age: {self.age}")
    print(f"Salary: {self.salary}")
  def __str__(self):
    return f"Employee(name={self.name}, surname={self.surname}, age={self.age}, salary={self.salary})"

  def __repr__(self):
    return self.__str__()  # Often __repr__ is the same as __str__

class EmployeeManager:
  _employees = []
  def __init__(self):
    self._employees = []

  def add_employee(employee):
    with open("manager.txt", "a") as file:
      file.write("\n")
      file.write(str(employee))
      file.close()
  def fire_employee_by_age(age_min, age_max):
      pattern = re.compile(r'age=(\d{2})')

      with open("manager.txt", "r+") as file:
          content = file.readlines()
          new_content = []
          fired = 0
          for line in content:
              match = pattern.findall(line)
              if match:
                  wiek = int(match[0])
                  if not (wiek >= age_min and wiek <= age_max):
                      new_content.append(line)
                      fired += 1
                  else:
                    pass
              else:
                pass
          file.seek(0)
          file.writelines(new_content)
          file.truncate()
          file.close()
          if fired > 0:
            print(f"Employees between {age_min} and {age_max} have been fired")
          else:
            print(f"No employees found between {age_min} and {age_max}")
  def display_employees():
    with open("manager.txt", "r") as file:
      for line in file:
          print(line)
      file.close()

  def find_employee_by_name(name_f,surname_f):
    with open("manager.txt", "r+") as file:
      for line in file:
        if name_f in line and surname_f in line:
          return line
      file.close()
  def update_employee_salary(name,surname, new_salary):
    with open("manager.txt", "r+") as file:
      content = file.readlines()
      emp = EmployeeManager.find_employee_by_name(name,surname)
      if not emp:
        return None
      else:
        for i in range(len(content)):
          if content[i] == emp:
            content[i] = re.sub(r"salary=(\d+)", f"salary={new_salary}", content[i])
        file.seek(0)
        file.writelines(content)
        file.close()



class FrontendManager(EmployeeManager):
  def __init__(self):
    super().__init__()
  def add_employee():
    name = input("Enter employee name: ")
    if not name:
      print("Name cannot be empty")
      return
    elif not name.isalpha():
      print("Name must contain only letters")
      return
    else:
      surname = input("Enter employee surname: ")
      if not surname:
        print("Surname cannot be empty")
        return
      elif not surname.isalpha():
        print("Surname must contain only letters")
        return
      else:
        age = int(input("Enter employee age: "))
        if not age:
          print("Age cannot be empty")
          return
        elif not 18 <= age <= 65:
          print("Age must be between 18 and 65")
          return
        elif not isinstance(age, int):
          print("Age must be an integer")
          return
        elif age.isalpha():
          print("Age cannot contain letters")
          return
        else:
          salary = int(input("Enter employee salary: "))
          if not salary:
            print("Salary cannot be empty")
            return
          elif not salary > 4200:
            print("Salary must at least match minimum wage")
            return
          elif salary.isalpha():
            print("Salary cannot contain letters")
            return
          else:
            new_employee = Employee(name,surname,age,salary)
            EmployeeManager.add_employee(new_employee)
            print("Employee has been added")
  def fire_employee_by_age():
    age_min = int(input("Enter minimum age: "))
    if not age_min:
      print("Minimum age cannot be empty")
      return
    elif not 18 <= age_min <= 65:
      print("Minimum age must be between 18 and 65")
      return
    elif not isinstance(age_min, int):
      print("Minimum age must be an integer")
      return
    elif age_min.isalpha():
      print("Minimum age cannot contain letters")
      return
    else:
      age_max = int(input("Enter maximum age: "))
      if not age_max:
        print("Maximum age cannot be empty")
        return
      elif not 18 <= age_max <= 65:
        print("Maximum age must be between 18 and 65")
        return
      elif not isinstance(age_max, int):
        print("Maximum age must be an integer")
        return
      elif age_max.isalpha():
        print("Maximum age cannot contain letters")
        return
      else:
        EmployeeManager.fire_employee_by_age(age_min,age_max)

  def update_employee_salary():
    name = input("Enter employee name: ")
    if not name:
      print("Name cannot be empty")
      return
    elif not name.isalpha():
      print("Name must contain only letters")
      return
    else:
      surname = input("Enter employee surname: ")
      if not surname:
        print("Surname cannot be empty")
        return
      elif not surname.isalpha():
        print("Surname must contain only letters")
        return
      else:
        new_salary = int(input("Enter new salary: "))
        if not new_salary:
          print("Salary cannot be empty")
          return
        elif not new_salary > 4200:
          print("Salary must at least match minimum wage")
          return
        elif new_salary.isalpha():
          print("Salary cannot contain letters")
          return
        else:
          EmployeeManager.update_employee_salary(name,surname,new_salary)








login_name = input("Enter your name: ")
login_surname = input("Enter your surname: ")
if login_name == "admin" and login_surname == "admin":
        print("Logged in successfully!")
        with open("manager.txt", "w") as file:
            file.seek(0)
            file.truncate()
            file.close()
        prac1 = Employee("John","Wayne", 25, 5000)
        prac2 = Employee("Jane", "Doe", 30, 6000)
        prac3 = Employee("Bob", "Ross", 35, 7000)

        EmployeeManager.add_employee(prac1)
        EmployeeManager.add_employee(prac2)
        EmployeeManager.add_employee(prac3)
        Opcja = 10
        while(Opcja != 0):
          print("Choose an option: ")
          print("1. Add employee")
          print("2. Fire employee by age")
          print("3. Display employees")
          print("4. Update employee salary")
          print("0. Exit")
          Opcja = int(input())
          match (Opcja):
            case 1:
              FrontendManager.add_employee()
            case 2:
              FrontendManager.fire_employee_by_age()
            case 3:
              FrontendManager.display_employees()
            case 4:
              FrontendManager.update_employee_salary()
            case 0:
              print("Exiting...")
              break
            case _:
              print("Invalid option")
else:
    print("Invalid login data. Exiting program...")
    exit()











```

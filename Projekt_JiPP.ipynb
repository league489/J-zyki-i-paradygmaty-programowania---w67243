```python
#wersja bardziej funkcjna
import heapq

class Zadanie:
    def __init__(self, nazwa, czas, kolejność=0, priorytet=0):
        self.nazwa = nazwa
        self.czas = czas
        self.priorytet = priorytet
        self.kolejność = kolejność

    def __lt__(self, other):
        if self.czas == other.czas:
            return self.kolejność < other.kolejność
        return self.czas < other.czas

    def __repr__(self):
        return f"({self.nazwa}, {self.czas}, {self.kolejność}, {self.priorytet})"

def dodaj_zadanie(Zadania):
    zadanie = input("Podaj zadanie i jego czas wykonania: ").split()
    nazwa = zadanie[0]
    czas = int(zadanie[1])
    kolejność = len(Zadania) + 1
    zadanie = Zadanie(nazwa, czas, kolejność)
    Zadania.append(zadanie)

def oblicz_czas_oczekiwania(Zadania):
    czas_oczekiwania = 0
    Całkowity_czas_oczekiwania = 0
    for z in Zadania:
        print(f"Czas oczekiwania procesu {z.nazwa} wynosi {czas_oczekiwania}")
        Całkowity_czas_oczekiwania += czas_oczekiwania
        czas_oczekiwania += z.czas
    return Całkowity_czas_oczekiwania

def main():
    print("Harmonogramowanie zadań")
    Zadania = []
    while input("Czy chcesz dodać kolejne zadanie? Tak/Nie: ") != "Nie":
        dodaj_zadanie(Zadania)

    print("Przed harmonogramowaniem")
    print(Zadania)
    Całkowity_czas_oczekiwania1 = oblicz_czas_oczekiwania(Zadania)
    Średni_czas_oczekiwania1 = Całkowity_czas_oczekiwania1 / len(Zadania)
    print(f"Całkowity czas oczekiwania wynosi {Całkowity_czas_oczekiwania1}")
    print(f"Średni czas oczekiwania wynosi {Średni_czas_oczekiwania1}")

    heapq.heapify(Zadania)

    print("Po harmonogramowaniu")
    print(Zadania)
    Całkowity_czas_oczekiwania2 = oblicz_czas_oczekiwania(Zadania)
    Średni_czas_oczekiwania2 = Całkowity_czas_oczekiwania2 / len(Zadania)
    print(f"Całkowity czas oczekiwania wynosi {Całkowity_czas_oczekiwania2}")
    print(f"Średni czas oczekiwania wynosi {Średni_czas_oczekiwania2}")

    print(f"\nDzięki zastosowaniu algorytmu SJF zoptymalizowano całkowity czas realizacji zadań o {Całkowity_czas_oczekiwania1 - Całkowity_czas_oczekiwania2}, a średni czas o {Średni_czas_oczekiwania1 - Średni_czas_oczekiwania2}")

if __name__ == "__main__":
    main()
```
